# RAG Demo (Baseline Example)

This notebook demonstrates a simple Retrieval-Augmented Generation (RAG) pipeline using the platform services.

This is a baseline implementation intended for learning and experimentation.

It does NOT include:
- advanced retrieval strategies
- reranking
- evaluation pipelines
- production optimizations

The goal is to show how external projects can integrate with:
- Qdrant (vector store)
- Ollama (LLM runtime)
- platform infrastructure

## RAG Flow

```text
Markdown docs → Chunking → Embeddings → Qdrant → Retrieval → Ollama → Grounded answer
```

The notebook is intentionally practical and lightweight so it can be run cell by cell during local platform testing.


## How to Run

1. Ensure Qdrant is running:
   .\scripts\windows\up-full.ps1

2. Ensure Ollama is running and model is available:
   ollama pull llama3.1:8b

3. Restart the notebook kernel

4. Run all cells from top to bottom

## Section 2: Embedding Model

Load the embedding model used to convert documents and queries into vector representations for semantic search.


````python
print(f"Loading embedding model: {EMBEDDING_MODEL_NAME}")

try:
    st_model = SentenceTransformer(EMBEDDING_MODEL_NAME, device="cpu")

    class LocalEmbeddings:
        """Simple wrapper to make SentenceTransformer compatible with LangChain."""

        def __init__(self, model):
            self.model = model

        def embed_documents(self, texts):
            return self.model.encode(
                texts,
                batch_size=32,
                normalize_embeddings=True,
                show_progress_bar=True
            ).tolist()

        def embed_query(self, text):
            return self.model.encode(
                [text],
                normalize_embeddings=True
            )[0].tolist()

    embeddings = LocalEmbeddings(st_model)

    EMBEDDING_DIM = st_model.get_sentence_embedding_dimension()

    print(f"Embedding model loaded: {EMBEDDING_MODEL_NAME}")
    print(f"Dimension: {EMBEDDING_DIM}")

    # sanity check
    sample = embeddings.embed_query("What is RAG?")
    print(f"Sample embedding length: {len(sample)}")

except Exception as e:
    print(f"Failed to load embedding model '{EMBEDDING_MODEL_NAME}': {e}")
    raise
````

## Section 3: Load and Chunk Documents

Load files from the selected source paths, filter supported content, and split them into smaller chunks for embedding and retrieval.

### Step 3.1 – Define Document Loader

Define a function to:
- scan configured source paths
- filter allowed file types
- exclude sensitive/system files
- split content into overlapping chunks
- attach metadata for traceability

````python
def load_docs(
    source_paths: list[str] = SOURCE_PATHS,
    chunk_size: int = CHUNK_SIZE,
    chunk_overlap: int = CHUNK_OVERLAP,
) -> list[Document]:
    docs = []

    def chunk_text(text: str, size: int, overlap: int) -> list[str]:
        """Split text into overlapping character chunks."""
        if not text or not text.strip():
            return []

        chunks = []
        step = max(1, size - overlap)

        for i in range(0, len(text), step):
            chunk = text[i:i + size].strip()
            if chunk:
                chunks.append(chunk)

            if i + size >= len(text):
                break

        return chunks

    def detect_source_type(file_path: Path) -> str:
        """Infer a simple source type from file path."""
        parts = {p.lower() for p in file_path.parts}
        name = file_path.name.lower()

        if name == "readme.md":
            return "readme"
        if "compose" in parts:
            return "compose"
        if "docs" in parts:
            return "doc"
        if "scripts" in parts:
            return "script"
        if "config" in parts:
            return "config"
        if name == "dockerfile" or name == ".dockerignore":
            return "docker"
        if file_path.suffix.lower() == ".ps1":
            return "setup_script"

        return "other"

    def is_allowed_file(file_path: Path) -> bool:
        """Allow only approved files and exclude sensitive ones."""
        filename = file_path.name

        if filename in EXCLUDED_FILE_NAMES:
            return False

        if filename in ALLOWED_FILE_NAMES:
            return True

        if file_path.suffix.lower() in ALLOWED_EXTENSIONS:
            return True

        return False

    def add_file(file_path: Path):
        """Read a file, chunk it, and add as LangChain Documents."""
        try:
            content = file_path.read_text(encoding="utf-8")
        except UnicodeDecodeError:
            print(f"Skipping non-text file: {file_path}")
            return
        except Exception as e:
            print(f"Failed to load {file_path}: {e}")
            return

        chunks = chunk_text(content, chunk_size, chunk_overlap)
        source_type = detect_source_type(file_path)

        for i, chunk in enumerate(chunks):
            docs.append(
                Document(
                    page_content=chunk,
                    metadata={
                        "source": file_path.name,
                        "path": str(file_path),
                        "type": source_type,
                        "chunk": i,
                    },
                )
            )

    for source in source_paths:
        path = Path(source)

        if not path.exists():
            print(f"Path not found, skipping: {path}")
            continue

        if path.is_file():
            if is_allowed_file(path):
                print(f"Loading file: {path}")
                add_file(path)
            else:
                print(f"Skipping unsupported file: {path}")
            continue

        if path.is_dir():
            print(f"Scanning directory: {path}")

            for root, dirs, files in os.walk(path):
                dirs[:] = [d for d in dirs if d not in EXCLUDED_DIR_NAMES]

                for file_name in files:
                    file_path = Path(root) / file_name

                    if is_allowed_file(file_path):
                        print(f"Loading file: {file_path}")
                        add_file(file_path)

    if not docs:
        print("No files found. Using demo fallback.")
        return [
            Document(
                page_content="Harmonia includes Qdrant, Ollama, MLflow, and Docker Compose orchestration.",
                metadata={
                    "source": "demo.txt",
                    "path": "fallback",
                    "type": "demo",
                    "chunk": 0,
                },
            )
        ]

    print(f"Loaded {len(docs)} chunks from configured sources.")
    return docs
````

### Step 3.2 – Load and Inspect Documents

Execute the loader to:
- build the document collection
- verify chunking quality
- inspect metadata structure

````python
documents = load_docs()
print(f"Total document chunks: {len(documents)}")

for doc in documents[:3]:
    print("\n--- Sample chunk ---")
    print(doc.page_content[:300])
    print(doc.metadata)
````

## Section 4: Create Vector Store and Index Documents

Create the Qdrant collection, convert document chunks into embeddings, and store them with metadata for semantic retrieval.

### Step 4.1 – Index Documents into Qdrant

Initialize the Qdrant vector store and store the prepared document chunks as embeddings in the configured collection.

````python
print(f"Creating vector store in collection: {COLLECTION_NAME}")

try:
    vector_store = QdrantVectorStore.from_documents(
        documents=documents,
        embedding=embeddings,
        url=f"http://{QDRANT_HOST}:{QDRANT_PORT}",
        collection_name=COLLECTION_NAME,
        force_recreate=True,
    )

    print("Vector store created successfully")
    print(f"Indexed {len(documents)} document chunks into '{COLLECTION_NAME}'")

except Exception as e:
    print(f"Failed to create vector store: {e}")
    raise
````

### Step 4.2 – Verify Indexed Collection

Check that the Qdrant collection exists and confirm the number of indexed vectors.

````python
info = qdrant_client.get_collection(COLLECTION_NAME)

print(f"Collection: {COLLECTION_NAME}")
print(f"Points count: {info.points_count}")
print(f"Status: {info.status}")
````

## Section 5: Create and Test the Retriever

Create a retriever from the indexed vector store and test semantic search before involving the LLM.

### Step 5.1 – Create the Retriever

Convert the indexed Qdrant vector store into a retriever that returns the top matching chunks for a query.

````python
retriever = vector_store.as_retriever(search_kwargs={"k": TOP_K})

print("Retriever created successfully")
print(f"Top-k: {TOP_K}")
````

### Step 5.2 – Test Retrieval

Run a sample query and inspect the retrieved chunks and metadata.

````python
test_query = "What components does Harmonia include?"

retrieved_docs = retriever.invoke(test_query)

print(f"Query: {test_query}")
print(f"Retrieved chunks: {len(retrieved_docs)}")

for i, doc in enumerate(retrieved_docs, 1):
    print(f"\n--- Retrieved chunk {i} ---")
    print(doc.page_content[:500])
    print("Metadata:", doc.metadata)
````

## Section 6: Connect the LLM and Build the Prompt

Initialize the local LLM and define the prompt template that combines retrieved context with the user question.


### Step 6.1 – Initialize the LLM

Create the Ollama chat model that will generate answers from retrieved context.

````python
llm = ChatOllama(
    model=MODEL_NAME,
    base_url=OLLAMA_HOST,
    temperature=0
)

print(f"LLM initialized: {MODEL_NAME}")
````

### Step 6.2 – Define the Prompt Template

Create a prompt that instructs the model to answer only from the retrieved context.

````python
prompt = ChatPromptTemplate.from_template("""
You are an assistant answering questions about Harmonia - Modular AI & Data Platform.

Use only the context below to answer the question.
If the answer is not in the context, say you do not know.

Context:
{context}

Question:
{question}
""")

print("Prompt template created")
````

## Section 7: Build the RAG Chain and Ask Questions

Combine the retriever, prompt, and LLM into a full RAG pipeline, then test it with example questions.

### Step 7.1 – Format Retrieved Documents

Create a helper function to combine retrieved chunks into a single context string.

````python
def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

print("Document formatting function ready")
````

### Step 7.2 – Build the RAG Chain

Connect retrieval, prompt construction, LLM generation, and output parsing into one pipeline.

````python
rag_chain = (
    {
        "context": retriever | format_docs,
        "question": RunnablePassthrough(),
    }
    | prompt
    | llm
    | StrOutputParser()
)

print("RAG chain created successfully")
````

### Step 7.3 – Test the Full RAG Pipeline

Run a sample question through the complete RAG system and inspect the final answer.

````python
question = "What are the main components of the platform architecture and their roles?"
answer = rag_chain.invoke(question)

print("Question:", question)
print("\nAnswer:\n")
print(answer)
````

````python
questions = [
    "What components does Harmonia include?",
    "How is orchestration handled in the platform?",
    "What services are defined in Docker Compose?",
    "How can the platform be started on Windows?",
]

for q in questions:
    print("\n" + "=" * 80)
    print("Question:", q)
    print("\nAnswer:\n")
    print(rag_chain.invoke(q))
````

## Section 8: Interactive RAG Chat

Run a simple interactive chatbot loop in the notebook to query the indexed knowledge base in real time.

### Step 8.1 – Start Interactive Chat

Run a simple notebook chat loop that sends questions through the RAG pipeline and displays the retrieved sources for inspection.

````python
print("RAG chatbot ready. Type 'exit' or 'quit' to stop.\n")

turn = 1

while True:
    try:
        question = input(f"You [{turn}]: ").strip()

        if question.lower() in {"exit", "quit"}:
            print("Chatbot stopped.")
            break

        if not question:
            print("Please enter a question.\n")
            continue

        retrieved_docs = retriever.invoke(question)
        answer = rag_chain.invoke(question)

        print("\nAssistant:")
        print(answer if answer else "[No answer returned]")

        print("\nRetrieved context:")
        if retrieved_docs:
            for i, doc in enumerate(retrieved_docs, 1):
                source = doc.metadata.get("source", "unknown")
                doc_type = doc.metadata.get("type", "unknown")
                chunk = doc.metadata.get("chunk", "unknown")
                preview = doc.page_content[:200].replace("\n", " ")
                print(f"\n[{i}] {source} | {doc_type} | chunk {chunk}")
                print(f"    {preview}...")
        else:
            print("No documents retrieved.")

        print("\n" + "-" * 80 + "\n")
        turn += 1

    except KeyboardInterrupt:
        print("\nChatbot interrupted by user.")
        break
    except Exception as e:
        print(f"\nError during chat: {e}")
        print("\n" + "-" * 80 + "\n")
````
